````python
import pandas as pd
# Подгрузка данных из csv-файла
LTV_dataset_raw = pd.read_csv("LTV.csv")
LTV_dataset_raw.tail()
````

````python
# Всего записей в датасете (строки, колонки)
print('All(rows, cols):', LTV_dataset_raw.shape)
````

````python
# Количество уникальных записей в "camera_model"
list_unique_model = LTV_dataset_raw.camera_model.unique()
print("camera_models:", len(list_unique_model))
# Дубли в колонке 'camera_model'
list_duplicates_model = LTV_dataset_raw[LTV_dataset_raw['camera_model'].duplicated()]
print("duplicates:\n", list_duplicates_model[['camera_model']])
````

````python
# Удаление из колонки 'camera_sensitivity_raw' слов "чувствительность" и "чувствительнось", удаление крайних пробелов
LTV_dataset_raw['camera_sensitivity_raw'] = LTV_dataset_raw['camera_sensitivity_raw'].str.replace(r'[чувствительность|чувствительнось]+', '', regex=True).str.strip()
#LTV_dataset_raw['camera_sensitivity_raw'].tail()
# Добавить колонку 'camera_sensitivity_day_x10000'
LTV_dataset_raw['camera_sensitivity_day_x10000'] = LTV_dataset_raw['camera_sensitivity_raw'].str.extract(r'^\s*(-?\d+\.\d+)').astype(float)*10000
LTV_dataset_raw['camera_sensitivity_day_x10000'] = LTV_dataset_raw['camera_sensitivity_day_x10000'].round().astype(int)
# Варианты значений чувствительности (camera_sensitivity_day_x10000)
list_unique_sensitivity = LTV_dataset_raw['camera_sensitivity_day_x10000'].unique()
for i in list_unique_sensitivity:
    print(f"__{i}__")
LTV_dataset_raw.head()
````

````python
# Варианты значений температуры (operating_temperature)
list_unique_operating_temperature = LTV_dataset_raw['camera_operating_temperature_raw'].unique()
for i in list_unique_operating_temperature:
    print(f"__{i}__")
````

````python
# Добавить колонки с минимальным и максимальных значением рабочих температур
LTV_dataset_raw['camera_operating_temperature_min'] = LTV_dataset_raw['camera_operating_temperature_raw'].str.extract(r'^\s*(-?\d+\.?\d*)')
LTV_dataset_raw['camera_operating_temperature_max'] = LTV_dataset_raw['camera_operating_temperature_raw'].str.extract(r'(-?\d+\.?\d*)\s*$')
````

````python
# Удалить "1x" из колонки с интерфейсами
LTV_dataset_raw['camera_data_transfer_interface']=LTV_dataset_raw['camera_data_transfer_interface'].str.replace('1x', '')
# Варианты значений интерфейса передачи данных (data_transfer_interface)
list_unique_data_transfer_interface = LTV_dataset_raw['camera_data_transfer_interface'].unique()
for i in list_unique_data_transfer_interface:
    print(f"__{i}__")
````

````python
# Варианты корпуса (case_type)
list_unique_case_type = LTV_dataset_raw['camera_case_type'].unique()
for i in list_unique_case_type:
    print(f"__{i}__")
# Не указан тип корпуса
nan_rows = LTV_dataset_raw[LTV_dataset_raw['camera_case_type'].isna()]
print('case_type is nan:\n', nan_rows[['camera_model']])
````

````python
# resolution
# Заменить в разрешении два пробела на один
LTV_dataset_raw['camera_resolution_raw'] = LTV_dataset_raw['camera_resolution_raw'].str.replace('  ', ' ')
# Варианты разрешения (resolution)
list_unique_resolution = LTV_dataset_raw['camera_resolution_raw'].unique()
for i in list_unique_resolution:
    print(f"__{i}__")
# Создать колонку camera_resolution
LTV_dataset_raw['camera_resolution'] = LTV_dataset_raw['camera_resolution_raw'].str.extract(r'^(\d+)\s*Mp$')
LTV_dataset_raw[['camera_resolution_raw','camera_resolution']].head()
````

````python
# power
# Варианты power
#list_unique_power = LTV_dataset_raw['camera_power_raw'].unique()
#for i in list_unique_power:
#    print(f"__{i}__")
# Создать колонку camera_power_x10
LTV_dataset_raw['camera_power_x10'] = LTV_dataset_raw['camera_power_raw'].str.replace(',', '.').str.extract(r'(\d+\.?\d*)\s*Вт').astype(float)*10
LTV_dataset_raw['camera_power_x10'] = LTV_dataset_raw['camera_power_x10'].round().astype(int)
#list_unique_power = LTV_dataset_raw['camera_power'].unique()
#for i in list_unique_power:
#    print(f"__{i}__")
LTV_dataset_raw[['camera_power_raw','camera_power_x10']].head()
# Не указана мощность
#nan_power = LTV_dataset_raw[LTV_dataset_raw['camera_power'].isna()]
#print('case_type is nan:\n', nan_rows[['camera_model']])
````

````python
# poe
# Создать колонку camera_power_supply_poe
LTV_dataset_raw['camera_power_supply_poe'] = LTV_dataset_raw['camera_power_supply_raw'].str.contains('poe', case=False, na=False)
#list_unique_power_supply_raw = LTV_dataset_raw['camera_power_supply_raw'].unique()
#for i in list_unique_power_supply_raw:
#    print(f"__{i}__")
LTV_dataset_raw[['camera_power_supply_raw','camera_power_supply_poe']].head()
````

````python
# IR_illumination
import re


list_unique_IR_illumination = LTV_dataset_raw['camera_IR_illumination'].unique()
for i in list_unique_IR_illumination:
    print(f"__{i}__")
# Удалить прочерк и пробелы с буквой м
LTV_dataset_raw['camera_IR_illumination'] = LTV_dataset_raw['camera_IR_illumination'].str.replace(' м', '').str.replace('-','0').str.replace('нет','0').apply(lambda x: re.sub(r'[^0-9]', '0', str(x)))
````

````python
# camera_lightning_protection
#list_unique_lightning_protection = LTV_dataset_raw['camera_lightning_protection'].unique()
#for i in list_unique_lightning_protection:
#    print(f"__{i}__")
LTV_dataset_raw['camera_lightning_protection'] = LTV_dataset_raw['camera_lightning_protection'].str.contains('Есть|4кВ|6кВ', case=False, na=False)
````

````python
# Сохранить датасет в файл
LTV_dataset_raw.to_csv("LTV_dataset.csv")
#LTV_dataset_raw[['camera_model','camera_resolution','camera_sensitivity_day','camera_IR_illumination','camera_case_type','camera_operating_temperature_raw']].head()
LTV_dataset_raw.head()
#LTV_dataset_raw.shape
````

Импортировать данные из csv-файла

#### YDB CLI, Cloud Shell (bash)

Подразумевается, что используется serverless YDB и Cloud Shell открыта в одном каталоге (фолдере) с базой. 

Создать файл (например, ```vim ltv.csv```) и скопипастить туда содержимое.

Первой строкой в файле должны идти названия колонок.

Заменить ```False``` на ```false```, а ```True``` на ```true```. Команды для Vim ```:%s/False/false/g``` ```:%s/True/true/g```

Импортировать данные из файла "ltv.csv" в таблицу "ltv". Для авторизации используются метаданные.

```bash
ydb \
    -e grpcs://ydb.serverless.yandexcloud.net:2135 \
    -d /ru-central1/b1go5m8qrt4s48isg7c1/etn84ja0b6rogn2aef00 \
    --use-metadata-credentials \
    import file csv -p ltv --header ltv.csv
```
